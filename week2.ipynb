```python
sales_data_copy = sales_data_copy.apply(lambda col: col.str.capitalize() if col.dtype == "object" else col)

product_info_copy = product_info_copy.apply(lambda col: col.str.capitalize() if col.dtype == "object" else col)

cust_info_copy = cust_info_copy.apply(lambda col: col.str.capitalize() if col.dtype == "object" else col)

cust_info_copy['loyalty_tier'] = (
    cust_info_copy['loyalty_tier']
    .astype(str)       
    .str.strip()        
    .str.replace(r'\s+', ' ', regex=True)  
    .str.capitalize()   
)
```

## Task 4

```python
merged_df['revenue'] = merged_df['quantity'] * merged_df['unit_price'] * (1 - merged_df['discount_applied'])

print(merged_df['revenue'])
```

```python
merged_df['order_date'] = pd.to_datetime(
    merged_df['order_date'], errors='coerce', dayfirst=True
)

merged_df['order_week'] = merged_df['order_date'].dt.isocalendar().week

print(merged_df[['order_date', 'order_week']])

```

```python
merged_df['price_band'] = pd.cut(
    merged_df['unit_price'],
    bins=[-float('inf'), 15, 30, float('inf')],
    labels=['Low (<15)', 'Medium (1530)', 'High (>30)']
)

print(merged_df[['unit_price', 'price_band']].head(10))
```

```python
merged_df['days_to_order'] = (merged_df['order_date'] - merged_df['launch_date']).dt.days

print(merged_df['days_to_order'].head(10))
```

```python
merged_df['email_domain'] = merged_df['email'].str.extract(r'@([\w\.-]+)')

print(merged_df['email_domain'].head(10))
```

```python
merged_df['is_late'] = merged_df['delivery_status'].str.lower().eq('delayed')

print(merged_df['is_late'].head(10))
```

```python
weekly_revenue = (
    merged_df
    .groupby(['region_cust', 'order_week'], as_index=False)
    .agg(total_revenue=('revenue', 'sum'))
    .sort_values(['region_cust', 'order_week'])
)

print("Weekly Revenue Trends by Region:")
print(weekly_revenue.to_string())
```

```python
category_perf = (
    merged_df
    .groupby('category', as_index=False)
    .agg(
        total_revenue=('revenue', 'sum'),
        total_quantity=('quantity', 'sum'),
        avg_discount=('discount_applied', 'mean')
    )
    .sort_values('total_revenue', ascending=False)
)

print("Product Category Performance:")
print(category_perf.to_string())
```

```python
merged_df['signup_date'] = pd.to_datetime(
    merged_df['signup_date'], errors='coerce', dayfirst=True
)


merged_df['signup_month'] = merged_df['signup_date'].dt.month

customer_behavior = (
    merged_df
    .groupby(['loyalty_tier', 'signup_month'], as_index=False)
    .agg(
        total_customers=('customer_id', 'nunique'),
        total_revenue=('revenue', 'sum'),
        avg_order_value=('revenue', 'mean')
    )
)

print("Customer Behaviour by Loyalty Tier and Signup Month:")
print(customer_behavior.to_string())
```

```python
delivery_perf = (
    merged_df
    .groupby(['region_cust', 'price_band'], as_index=False)
    .agg(
        total_orders=('order_id', 'count'),
        delayed_orders=('is_late', 'sum'),
        delay_rate=('is_late', 'mean')
    )
    .sort_values('delay_rate', ascending=False)
)

print("Delivery Performance by Region and Price Band:")
print(delivery_perf.to_string())
```

```python
print(merged_df.columns.tolist())
```

```python
payment_pref = (
    merged_df
    .groupby(['loyalty_tier', 'payment_method'], as_index=False)
    .agg(num_transactions=('order_id', 'count'))
    .sort_values(['loyalty_tier', 'num_transactions'], ascending=[True, False])
)

print("Preferred Payment Methods by Loyalty Tier:")
print(payment_pref.to_string())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.lineplot(
    data=merged_df,
    x='order_week',
    y='revenue',
    hue='region_cust',
    marker='o'
)
plt.title("Weekly Revenue Trends by Region")
plt.xlabel("ISO Week Number")
plt.ylabel("Revenue ()")
plt.legend(title="Region")
plt.tight_layout()
plt.show()
```

```python
top_categories = (
    merged_df.groupby('category')['revenue']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

plt.figure(figsize=(8,6))
sns.barplot(
    data=top_categories, 
    x='revenue', 
    y='category'
)
plt.title("Top 5 Categories by Total Revenue")
plt.xlabel("Revenue ()")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(
    data=merged_df,
    x='category',
    y='quantity',
    hue='discount_applied'
)
plt.title("Quantity Distribution by Category and Discount Applied")
plt.xlabel("Product Category")
plt.ylabel("Quantity Ordered")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
corr_cols = merged_df[['revenue', 'discount_applied', 'quantity']]

plt.figure(figsize=(6,5))
sns.heatmap(
    corr_cols.corr(),
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)
plt.title("Correlation Heatmap: Revenue, Discount, Quantity")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(
    data=merged_df,
    x='loyalty_tier',
    hue='region_cust'
)
plt.title("Order Count by Loyalty Tier and Region")
plt.xlabel("Loyalty Tier")
plt.ylabel("Number of Orders")
plt.xticks(rotation=30)
plt.legend(title="Region")
plt.tight_layout()
plt.show()
```

```python
delivery_stacked = merged_df.groupby(['price_band', 'delivery_status']).size().unstack(fill_value=0)

delivery_stacked.plot(
    kind='bar',
    stacked=True,
    figsize=(10,6)
)

plt.title("Delivery Status by Price Band")
plt.xlabel("Price Band")
plt.ylabel("Number of Orders")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
